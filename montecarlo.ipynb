<h2>Monte Carlo Simulation to Estimate &pi;</h2>

<h4>Theoretical Explanation</h4>

<ul>
  <li>
    <b>Setup:</b><br>
    Imagine a square with sides of length 1. Inside this square, draw a circle with radius 0.5 centered at (0.5, 0.5).
  </li>
  <li>
    <b>Random Sampling:</b><br>
    We generate random points inside the square.
  </li>
  <li>
    <b>Circle Check:</b><br>
    For each point, we check if it falls inside the circle using the formula:<br>
    <br>
    <code>(x - 0.5)<sup>2</sup> + (y - 0.5)<sup>2</sup> &le; 0.25</code>
    <br>
  </li>
  <li>
    <b>Estimating &pi;:</b><br>
    The probability that a random point falls inside the circle is equal to the ratio of the circle's area to the square's area:
    <ul>
      <li>Area of circle = &pi; &times; (0.5)<sup>2</sup> = <sup>&pi;</sup>&frasl;<sub>4</sub></li>
      <li>Area of square = 1</li>
    </ul>
    So,<br>
    <br>
    <span style="font-size:1.2em;">
      <sup>Points inside circle</sup>&frasl;<sub>Total points</sub> &asymp; <sup>&pi;</sup>&frasl;<sub>4</sub>
    </span>
    <br><br>
    Therefore,<br>
    <br>
    <span style="font-size:1.2em;">
      &pi; &asymp; 4 &times; <sup>Points inside circle</sup>&frasl;<sub>Total points</sub>
    </span>
    <br>
  </li>
</ul>





```python
import random
import matplotlib.pyplot as plt

# Simulation parameters
num_points = 1000  # Try increasing this for better accuracy
inside = 0

# Generate random points and check circle inclusion
x_inside, y_inside = [], []
x_outside, y_outside = [], []

for _ in range(num_points):
    x, y = random.random(), random.random()
    if (x-0.5)**2 + (y-0.5)**2 <= 0.25:  # Check if inside circle
        inside += 1
        x_inside.append(x)
        y_inside.append(y)
    else:
        x_outside.append(x)
        y_outside.append(y)

# Calculate π estimate
pi_estimate = 4 * inside / num_points

# Visualization
plt.figure(figsize=(6,6))
plt.scatter(x_inside, y_inside, color='blue', s=1)
plt.scatter(x_outside, y_outside, color='red', s=1)
plt.title(f'π ≈ {pi_estimate:.4f} (Actual: 3.1415...)')
plt.show()
```


<li>This simple simulation shows how random sampling can help estimate mathematical constants like &pi;.</li>
<li><b>The more points you use, the more accurate your estimate becomes.</b></li>
<li>The method visually and mathematically demonstrates how probability and geometry are connected.</li>
<li>Monte Carlo simulations are powerful tools for solving problems that are hard to calculate directly, using randomness and statistics.</li>
